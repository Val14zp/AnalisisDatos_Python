# Crímenes contra personas físicas: Análisis de la criminalidad en Costa Rica en el periodo 2019 - 2024

## Cargando datos


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df: any = pd.read_csv("datos_criminalidad_2019-2024.csv", encoding='latin-1')
data: any = pd.DataFrame(df)
```

## Utilizando la función `count`


```python
cantidad_victimas_hombre: int = data.query("Genero == 'HOMBRE'")["Genero"].count()
cantidad_victimas_mujer: int = data.query("Genero == 'MUJER'")["Genero"].count()
cantidad_victimas_na: int = data.query("Genero == 'DESCONOCIDO'")["Genero"].count()
cantidad_victimas_peaton: int = data.query("SubVictima == 'PEATON [PERSONA]'")[
    "SubVictima"
].count()
cantidad_victimas_otro: int = data.query(
    "SubVictima == 'OTRO O INDETERMINADO [PERSONA]'"
)["SubVictima"].count()
cantidad_victimas_empresario: int = data.query(
    "SubVictima == 'EMPRESARIO/COMERCIANTE [PERSONA]'"
)["SubVictima"].count()
cantidad_victimas_menordeedad: int = data.query(
    "SubVictima == 'MENOR DE EDAD [PERSONA]'"
)["SubVictima"].count()
cantidad_victimas_taxista: int = data.query("SubVictima == 'TAXISTA [PERSONA]'")[
    "SubVictima"
].count()
cantidad_victimas_estudiante: int = data.query("SubVictima == 'ESTUDIANTE [PERSONA]'")[
    "SubVictima"
].count()
cantidad_victimas_turista: int = data.query(
    "SubVictima == 'TURISTA/NACIONAL [PERSONA]'"
)["SubVictima"].count()
cantidad_victimas_pasajero: int = data.query(
    "SubVictima == 'PASAJERO DE TRANSPORTE PUBLICO [PERSONA]'"
)["SubVictima"].count()
cantidad_victimas_cliente: int = data.query(
    "SubVictima == 'CLIENTE LOCAL COMERCIAL [PERSONA]'"
)["SubVictima"].count()

cantidad_victimas_guanacaste: int = data.query("Provincia == 'GUANACASTE'")[
    "Provincia"
].count()
cantidad_victimas_puntarenas: int = data.query("Provincia == 'PUNTARENAS'")[
    "Provincia"
].count()
cantidad_victimas_alajuela: int = data.query("Provincia == 'ALAJUELA'")[
    "Provincia"
].count()
cantidad_victimas_heredia: int = data.query("Provincia == 'HEREDIA'")[
    "Provincia"
].count()
cantidad_victimas_sanjose: int = data.query("Provincia == 'SAN JOSE'")[
    "Provincia"
].count()
cantidad_victimas_cartago: int = data.query("Provincia == 'CARTAGO'")[
    "Provincia"
].count()
cantidad_victimas_limon: int = data.query("Provincia == 'LIMON'")["Provincia"].count()

print("Cantidad de víctimas por género:")
print("Hombres:", cantidad_victimas_hombre)
print("Mujeres:", cantidad_victimas_mujer)
print("Desconocido:", cantidad_victimas_na)


print("\nCantidad de víctimas por categoría:")
print("Peatones:", cantidad_victimas_peaton)
print("Otros o indeterminados:", cantidad_victimas_otro)
print("Empresarios o comerciantes:", cantidad_victimas_empresario)
print("Menores de edad:", cantidad_victimas_menordeedad)
print("Taxistas:", cantidad_victimas_taxista)
print("Estudiantes:", cantidad_victimas_estudiante)
print("Turistas:", cantidad_victimas_turista)
print("Pasajeros de transporte público:", cantidad_victimas_pasajero)
print("Clientes locales comerciales:", cantidad_victimas_cliente)

print("\nCantidad de víctimas por provincia:")
print("Guanacaste:", cantidad_victimas_guanacaste)
print("Puntarenas:", cantidad_victimas_puntarenas)
print("Alajuela:", cantidad_victimas_alajuela)
print("Heredia:", cantidad_victimas_heredia)
print("San José:", cantidad_victimas_sanjose)
print("Cartago:", cantidad_victimas_cartago)
print("Limón:", cantidad_victimas_limon)
```

## Utilizando `min` y `max`


```python
cantidad_de_crimenes_por_provincia: any = (
    data.query("Provincia != 'DESCONOCIDO'").groupby("Provincia")["Provincia"].count()
)



provincia_menos_crimenes: any = (
    cantidad_de_crimenes_por_provincia.idxmin(),
    cantidad_de_crimenes_por_provincia.min(),
)



provincia_mas_crimenes: any = (
    cantidad_de_crimenes_por_provincia.idxmax(),
    cantidad_de_crimenes_por_provincia.max(),
)



cantidad_por_genero: any = data["Genero"].value_counts()



genero_mas_comun: any = (
    cantidad_por_genero.idxmax(),
    cantidad_por_genero.max(),
)



genero_menos_comun: any = (
    cantidad_por_genero.idxmin(),
    cantidad_por_genero.min(),
)

print(cantidad_de_crimenes_por_provincia)

print(provincia_menos_crimenes)
print(provincia_mas_crimenes)

print(genero_mas_comun)
print(genero_menos_comun)
```

# Mapa de Calor de Análisis de Delitos por Provincia Durante el Año 2023, por Mes







```python
df_2023 = df[df['Año'] == 2023]
df_filtrado = df_2023[df_2023['Provincia'] != 'DESCONOCIDO']
mapa_calor_delitos_por_provincia = pd.crosstab(df_filtrado['Provincia'], df_filtrado['Mes'])
mapa_calor_delitos_por_provincia = mapa_calor_delitos_por_provincia.sort_values(by='Provincia', ascending=False)
plt.figure(figsize=(20, 7))
sns.heatmap(mapa_calor_delitos_por_provincia, annot=True, cmap= sns.cubehelix_palette(as_cmap=True), fmt='g')

plt.title("Análisis de Delitos por Provincia Durante el Año 2023, por Mes")
plt.show()
```

#Gráfico de barras 3D con el Análisis de la Cantidad de Víctimas por Género y grupo Etario

```python
df_grupo_etario_filtrado = df.query("Edad != 'Desconocido' and Edad != 'Adulto Mayor'")
df_genero_filtrado = df[df['Genero'] != 'DESCONOCIDO']

datos = pd.crosstab(df_grupo_etario_filtrado['Edad'], df_genero_filtrado['Genero'])


colores = ['#3C2750', '#D5ADB0']

fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111, projection='3d')

xpos = np.arange(len(datos.index))
ypos = np.arange(len(datos.columns))
xpos, ypos = np.meshgrid(xpos, ypos, indexing="ij")
xpos = xpos.ravel()
ypos = ypos.ravel()
zpos = np.zeros_like(xpos)


dx = dy = 0.75
dz = datos.values.ravel()

for i in range(len(datos.values.ravel())):
    ax.bar3d(xpos[i], ypos[i], zpos[i], dx, dy, dz[i], color=colores[i % len(colores)])

ax.set_xticks(np.arange(len(datos.index)))
ax.set_xticklabels(datos.index)
ax.set_yticks(np.arange(len(datos.columns)))
ax.set_yticklabels(datos.columns)
ax.set_xlabel('Grupo Etario')
ax.set_ylabel('Género')
ax.set_zlabel('Cantidad')
ax.set_title('Análisis de Grupo Etario por Género')

plt.show()
```
