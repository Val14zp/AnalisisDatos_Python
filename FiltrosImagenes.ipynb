```python

# Abrimos la imagen
imagen = io.imread ("Heredia.png")
imagen_g = rgb2gray (imagen)

# Filtros: sobel, roberts, prewitt
filtros = [filters.sobel, filters. roberts, filters.prewitt, filters.gaussian, filters.butterworth, filters.threshold_niblack] #Se encontraron otros tres filtros y se agregaron a la lista, el primero se nota el cambio de color morado a verde con amarillo y el siguiente lo pone totalmente verde. Luego tambin se evidencia en el ltimo filtro aplicado que lo que hace es poner la imagen borrosa.

for filtro in filtros:
# Aplicamos cada uno de los filtros
  img_fil = filtro(imagen_g)

  # Mostramos los resultados
  plt.imshow (img_fil)
  plt. show ()
```
