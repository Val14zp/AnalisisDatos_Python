```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df: any = pd.read_csv("datos_criminalidad_2019-2024.csv", encoding='latin-1')
data: any = pd.DataFrame(df)

delitos_por_mes = df.groupby(['Mes', 'Año']).size().reset_index(name='NumDelitos')

X = delitos_por_mes[['Mes', 'Año']]
y = delitos_por_mes['NumDelitos']


model = LinearRegression()
model.fit(X, y)


print(f'Intercepto: {model.intercept_}')
print(f'Coeficiente para Mes: {model.coef_[0]}')
print(f'Coeficiente para Año: {model.coef_[1]}')


meses = []
predicciones = []

print("\nPredicciones para los meses restantes del año 2024:")
for mes in range(4, 13):
  prediccion =  model.intercept_ + model.coef_[0] * mes + model.coef_[1] * 2024
  predicciones.append(prediccion)
  print(f"Mes {mes}, Año 2024: {prediccion:.2f} delitos")
  meses.append(mes)

plt.figure(figsize=(10, 6))
plt.plot(meses, predicciones, color='blue',  marker='o', linestyle='-', label='Predicciones para 2024')
plt.title('Predicciones de Delitos para el Año 2024')
plt.xlabel('Mes')
plt.ylabel('Número de Delitos')
plt.xticks(range(4, 13))
2
plt.legend()
plt.grid(True)
plt.show()

```
