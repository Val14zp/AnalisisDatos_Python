```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv("casasboston.csv")
```

```python
print(datos)
```

```python
df = datos[["RM","CRIM","MEDV","TOWN","CHAS"]]
```

```python
df=datos.rename(columns={
    "TOWN":"CIUDAD",
    "CRIM":"INDICE_CRIMEN",
    "INDUS":"PCT_ZONA_INDUSTRIAL",
    "CHAS": "RIO CHARLES",
    "RM": "N_HABITACIONES_MEDIO",
    "MEDV": "VALOR_MEDIANO",
    "LSTAT": "PCT_CLASE_BAJA"
})
```

```python
print(df)
```

```python
p= df.N_HABITACIONES_MEDIO.plot.hist()
plt.ylabel("Frecuencia")
plt.xlabel("Número de Habitaciones")
plt.title("Frecuencia del Número de Habitaciones")
plt.show()
```

```python
df.plot.scatter(x="VALOR_MEDIANO", y="INDICE_CRIMEN",alpha=0.5)
plt.title("Disperción del Índice de Criminalidad por Valor Mediano")
plt.xlabel("Valor Mediano")
plt.ylabel(" Índice de criminalidad")
plt.show()
```

```python
valor_por_ciudad=df.groupby("CIUDAD")["VALOR_MEDIANO"].mean()
valor_por_ciudad_ordenado = valor_por_ciudad.sort_values( ascending=True)
```

```python
valor_por_ciudad_ordenado.head(10).plot.barh()

plt.title("TOP 10 de Ciudades con Mayor Índice de Criminalidad")
plt.xlabel("Valor Mediano")
plt.ylabel("Ciudad")
plt.show()
```
